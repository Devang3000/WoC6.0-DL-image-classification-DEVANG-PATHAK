```python
# Q6

def remove_outliers_iqr(df, threshold=1.5):
    # Calculate the IQR for each numerical column
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1

    # Define a boolean mask for outliers
    outliers_mask = ((df < (Q1 - threshold * IQR)) | (df > (Q3 + threshold * IQR)))

    # Remove outliers from the DataFrame
    df_no_outliers = df[~outliers_mask]

    return df_no_outliers

# Identify and remove outliers for the YouTube DataFrame
df_youtube_no_outliers = remove_outliers_iqr(df_youtube)

# Identify and remove outliers for the Car DataFrame
df_car_no_outliers = remove_outliers_iqr(df_car)

# Identify and remove outliers for the Weather DataFrame
df_weather_no_outliers = remove_outliers_iqr(df_weather)

# Display the shape of the DataFrames before and after removing outliers
print("YouTube DataFrame shape before outliers:", df_youtube.shape)
print("YouTube DataFrame shape after outliers:", df_youtube_no_outliers.shape)

print("\nCar DataFrame shape before outliers:", df_car.shape)
print("Car DataFrame shape after outliers:", df_car_no_outliers.shape)

print("\nWeather DataFrame shape before outliers:", df_weather.shape)
print("Weather DataFrame shape after outliers:", df_weather_no_outliers.shape)
```
